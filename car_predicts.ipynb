```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2) 
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
# Hyperparameters

#Number of trees in RandomForest
n_estimators = [int(x)
                for x in np.linspace
                (start=100, stop= 120, num=12)]
print(n_estimators)

#Number of max no of feature to consider at every split
max_features = ['auto', 'sqrt']
print(max_features)

#max No of level in Trees
max_depth = [int(x)
                for x in np.linspace
                (5,30 , num=12)]
#max depth append.none
print(max_depth)

#min no of samples required to split a node
min_samples_split = [2,5,10, 15, 100]
print(min_samples_split)

#min no of samples required to at each leaf node
min_samples_leaf = [1, 2,5,10]
print(min_samples_leaf)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {
    'n_estimators':n_estimators,
    'max_features': max_features,
    'max_depth':max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf':min_samples_leaf
}

print(random_grid)
```

```python
#use RandonForestRegressor to search for best hyper_parameter

rfr = RandomForestRegressor()
```

```python
rfr_random = RandomizedSearchCV(estimator = rfr,
                         param_distributions = random_grid,scoring='neg_mean_squared_error', 
                         n_iter = 10, cv= 5, verbose=2, random_state=42, n_jobs=1)
```

```python
rfr_random.fit(x_train, y_train)
```

```python
prediction = rfr_random.predict(x_test)
prediction
```

```python
sns.displot(y_test-prediction)
```
